# Problem 1

Using matplotlib, plot the two points (x1,y1)=(3,5) and (x2,y2)=(7,2) and the line that passes through both points. The line must continue to the edges of the plot. (I.e., it is not just a line segment connecting the points.)

Your plot should demonstrate good style, as discussed in class, making it clear to read. 

```python
# Create the point variable
x1, y1 = 3, 5
x2, y2 = 7, 2
```

```python
# Calculate the slope and intercept of the line passing through the points
slope = (y2 - y1) / (x2 - x1)
intercept = y1 - slope * x1
```

```python
#Creating a plot

# Creating X and Y lines with 10 points gap so it goes outside the plot
x_line = [min(x1, x2) - 10, max(x1, x2) + 10]
y_line = [slope * x + intercept for x in x_line]

# Plot the points and the line
plt.plot(x_line, y_line, label='Line passing through points', linewidth=2, color='black')
plt.scatter([x1, x2], [y1, y2], color='black', marker='o', s=100, label='Points')

# plotting coordinates for point annotations
plt.text(x1, y1, (3,5), fontsize=14, verticalalignment='top', horizontalalignment='right')
plt.text(x2, y2, (7,2), fontsize=14, verticalalignment='top', horizontalalignment='right')

# Set and format axis labels and ticks
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Setting plot limits according to the provided data
plt.xlim(0,10)
plt.ylim(0,10)

# Add a coordinate grid
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Problem 2

Load data from the following parquet file (URL below) into Python. Make nice plots presenting the data as fully as possible. 
To fully present the data, you may need to make multiple plots.


```python
# Load the Parquet file into a DataFrame using Pandas
file = "exercises_problem2.parquet"
df_data = pd.read_parquet(file)
```

```python
#Preview data to check the visuals that can be made
df_data.head()
```

It can be seen that there are three variable columns with different float values, and one class column showing the categorizing them.

### Now first we will visualize the data in each column according to the class.

## VAR1

```python
# Creating Boxplot for comparing distrubution of Var1 Column with respect to Class Categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='class', y='var1', data=df_data)
plt.title(f'Boxplot for var1')
plt.xlabel("Class")
plt.ylabel('var1')
plt.show()
```

## VAR2

```python
# Creating Boxplot for comparing distrubution of var2 Column with respect to Class Categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='class', y='var2', data=df_data)
plt.title(f'Boxplot for var2')
plt.xlabel("Class")
plt.ylabel('var2')
plt.show()
```

## VAR3

```python
# Creating Boxplot for comparing distrubution of var3 Column with respect to Class Categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='class', y='var3', data=df_data)
plt.title(f'Boxplot for var3')
plt.xlabel("Class")
plt.ylabel('var3')
plt.show()
```

### Now we will create relational visuals for visualizing relationships between different columns and color coding them according to the class

```python
# Creating Pairplot visual from seaborn library
sns.pairplot(df_data, hue='class', palette='viridis', markers=["o", "o"])
plt.suptitle("Pairplot Showing Relations with Class Labels", y=1.02)
plt.show()
```
